##### Task 3.1: Load the contries.csv

```python
import pandas as pd
print("loading countries file ...")

data_frame = pd.read_csv("https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/countries.csv")
print("... resulting data frame loaded from the countries.csv:")
data_frame
```

---------------
##### Task 3.2: Basic Information & Statistics

```python
print("Number of (Rows, Columns) = ", data_frame.shape)
data_frame.describe()
```

##### Task 3.3: The last 4 rows

```python
df_last4 = data_frame.tail(4)
df_last4
```

---------------
##### Task 3.4: All Countries having € as currency

```python
df_euros = data_frame.loc[data_frame['Currency'] == 'EUR']

print("The following countries have the currency: €")
df_euros
```

---------------
##### Task 3.5: New Data Frame containing [Name, Currency]

```python
print("The new data frame contains:")
sub_df = data_frame[["Name", "Currency"]]
sub_df
```

---------------
##### Task 3.6: Countries having more than 2000 BIP

```python
print("The following countries have more than 2000 BIP")

df_bip2000 = data_frame.loc[data_frame['BIP'] > 2000]
df_bip2000
```

---------------
##### Task 3.7: Countries having inhabitants between 50 and 150 Mio.

```python
print("The following countries have a total of "
      "inhabitants between 50 and 150 million people:")
df_inhabitants = data_frame[
    data_frame['People'].between(50000000, 150000000, inclusive=False)
]
df_inhabitants
```

---------------
##### Task 3.8: Column Change from BIP to Bip

```python
print('New table with updated column name')
df_bip = data_frame.rename(columns={'BIP': 'Bip'})
df_bip
```

---------------
##### Task 3.9: Bip Sum Calculation

```python
bip_sum = df_bip['Bip'].sum()
print('The Bip sum over all countries is:\n', bip_sum, "Mrd. €")
```

---------------
##### Task 3.10: Average People Calculation

```python
mean_people = data_frame['People'].mean()
print('The average count of people over all countries is:\n', mean_people)
```

---------------
##### Task 3.11: Sort by Name in Ascending Order

```python
df_asc_sorted = data_frame.sort_values('Name', ascending=True)

print("Countries in ascending order:")
df_asc_sorted
```

---------------
##### Task 3.12: SMALL or BIG Country Evaluation

```python
import numpy as np
#df_area_cat = pd.DataFrame(data_frame, columns=['Name', 'People', 'Area', 'BIP', 'Currency'])
df_area_cat = data_frame.copy()
df_area_cat['Area'] = np.where(df_area_cat['Area'] <= 1000000, 'SMALL', 'BIG')
df_area_cat
```
